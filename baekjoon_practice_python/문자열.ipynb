### 아스키코드 (American Standard Code for Information Interchange)
영어 알파벳, 숫자, 특수문자 등을 컴퓨터에서 사용하기 위해 각각의 문자에 0~127 범위의 숫자를 할당함

##### ord() 함수
지정한 문자에 해당하는 아스키코드 값을 반환

```python
ord('A') #65
ord('a') #97
```

##### chr() 함수
지정한 아스키코드 값에 해당하는 문자를 반환

```python
chr(65) #A
chr(97) #a
```
